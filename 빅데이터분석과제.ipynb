```r
#3.2
?mpg
#drv is the type of drive train으로 f = front-wheel drive와 r = rear wheel drive, 4=4wd로 나뉜다.
library(tidyverse)
ggplot(data = mpg, mapping = aes(x = hwy, y = cyl)) +
  geom_point(aes(size=3, color="light purple"), alpha = 1/3)
```

```r
#3.3
#countinuous variable
library(tidyverse)
ggplot(data = mpg, aes(x = displ, y = cty)) +
  geom_point(aes(color = hwy, shape = trans, size = hwy), alpha = 1/3)

###color
#1. continuous: 연속적인 값 간의 차이 등을 비교 가능
#2. categorical: 서로 다른 category간의 구분 가능
###size
#1. continuous: 변수가 커지면 size도 커지고, 작아지면 size도 작아짐, 시각적으로 변수의 크기 차이를 가늠할 수 있음
#2. categorical: 범주형 변수인 경우 그 크기가 커진다고 해서 값이 커지는 것이 아니기에 잘 쓰이지는 않음
###shape
#1. continuous: 연속형 변수를 모양으로 매핑하는 것은 의미가 없음, 하고 싶다면 범주형 변수로 변환한 후 사용해야함
#2. categorical: 각 범주간의 차이를 시각적으로 확인할 수 있음
```

```r
#3.5
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ .)
##drv를 행으로 놓고, 열은 지정하지 않은 상태로 x축 displ로 y축을 hwy로 하는 그래프 산점도 생성
##drv를 행으로 하는 열이 지정되지 않은 서브플롯 생성
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(. ~ cyl)
##cyl를 열로 놓고, 행을 지정하지 않은 상태로 x축과 y축을 각각 displ, hwy로 하는 산점도 생성
##cyl를 열로 하는 행이 지정되지 않은 서브플롯 생성

##. 의 의미 = 해당 변수를 지정하지 않고, 기재해둔 변수를 기준으로 서브플롯을 만들라는 의미이다.
```
